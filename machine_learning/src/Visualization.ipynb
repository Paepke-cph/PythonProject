## Imports and Collect data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = pd.read_csv('cars.csv')
```

##  Basic Dataframe visualization

```python
display(dataset.shape)
display(dataset.columns)
```

## Cars: Make

```python
dups_make = dataset.pivot_table(index=['make'], aggfunc='size').reset_index()
plt.barh(dups_make.iloc[:,0], dups_make.iloc[:,1])
#plt.xticks(rotation=90)
```

```python
mean_year_make = dataset[['make','year','newprice']].groupby(['year','make']).mean().reset_index()
display(mean_year_make)
alfa_romeo = mean_year_make[mean_year_make['make'] == 'Alfa Romeo']
fiat = mean_year_make[mean_year_make['make'] == 'Fiat']
honda = mean_year_make[mean_year_make['make'] == 'Honda']

plt.plot(alfa_romeo['year'], alfa_romeo['newprice'])
plt.plot(fiat['year'], fiat['newprice'])
plt.plot(honda['year'], honda['newprice'])
```
